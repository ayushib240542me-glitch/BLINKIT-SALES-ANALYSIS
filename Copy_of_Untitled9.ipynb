```python
import pandas as pd

df = pd.read_csv('/content/blinkit_data (1).csv')

# Display the first 5 rows to confirm it loaded correctly
display(df.head(10))
```

```python
print(df['Item Fat Content'].unique())
```

```python
df['Item Fat Content']=df['Item Fat Content'].replace({'LF':'Low Fat','low fat':'Low Fat','reg':'Regular'})
```

```python
print(df['Item Fat Content'].unique())
```

# KPI REQUIREMENTS

```python
total_sales=df['Sales'].sum()
avg_sales=df['Sales'].mean()
no_of_items_sold=df['Sales'].count()
avg_ratings=df['Rating'].mean()
print(f"Total sales: ${total_sales:,.1f}")
print(f"Average sales: ${avg_sales}")
print(f"Number of items sold: {no_of_items_sold}")
print(f"Average rating: {avg_ratings:,.1f}")
```

### **CHARTREQUIREMENTS**

```python
sales_by_fat=df.groupby('Item Fat Content')['Sales'].sum()
plt.pie(sales_by_fat, labels=sales_by_fat.index, autopct='%.1f%%', startangle=90)
plt.title('Sale by Fat Content')
plt.axis('equal')
plt.show()
```

```python
sales_by_type=df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(10,6))
bars=plt.bar(sales_by_type.index,sales_by_type.values)
plt.title('Sales by Item Type')
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.xticks(rotation=-90)

for bar in bars:
  plt.text(bar.get_x()+bar.get_width()/2,bar.get_height(),f'{bar.get_height():,.1f}',ha='center',va='bottom',fontsize=8)
plt.tight_layout()
plt.show()
```

```python
grouped=df.groupby(['Outlet Location Type','Item Fat Content'])['Sales'].sum().unstack()
grouped=grouped[['Regular','Low Fat']]
ax=grouped.plot(kind='bar',figsize=(8,5), title='outlet tier by item fat content')
plt.xlabel('Outlet Location Type')
plt.ylabel('Total Sales')
plt.legend(title='Item Fat Content')
plt.tight_layout()
plt.show()
```

## **TOTAL SALES BY OUTLET ESTABLISHMENT**

```python
sales_by_year=df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()
plt.figure(figsize=(9,5))
plt.plot(sales_by_year.index,sales_by_year.values,marker='o',linestyle='-')
plt.title('Total Sales by Outlet Establishment Year')
plt.xlabel('Outlet Establishment Year')
plt.ylabel('Total Sales')
for x,y in zip(sales_by_year.index,sales_by_year.values):
  plt.text(x, y,f'{y:,.0f}',ha='center',va='bottom',fontsize=8)
plt.tight_layout()
plt.show()
```

# SALES BY OUTLET **SIZE**

```python
sales_by_size=df.groupby('Outlet Size')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(4,4))
plt.pie(sales_by_size,labels=sales_by_size.index,autopct='%.1f%%',startangle=90)
plt.title('Sales by Outlet Size')
plt.tight_layout()
plt.show()
```

# TOTAL SALES BY OUTLET LOCATION TYPE

```python
sales_by_location=df.groupby('Outlet Location Type')['Sales'].sum().reset_index()
sales_by_location=sales_by_location.sort_values(by='Sales',ascending=False)
plt.figure(figsize=(8,3))
ax=sns.barplot(x='Outlet Location Type',y='Sales',data=sales_by_location)
plt.title('Total Sales by Outlet Location Type')
plt.xlabel('Outlet Location Type')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```
